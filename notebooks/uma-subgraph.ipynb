| interpretation                | `payouts` you pass | denominator stored       | result                                 |
| ----------------------------- | ------------------ | ------------------------ | -------------------------------------- |
| **Yes** wins                  | `[1, 0]`           | 1                        | Yesshare  1 collateral; Noshare  0 |
| **No** wins                   | `[0, 1]`           | 1                        | Noshare  1 collateral; Yesshare  0 |
| **Invalid / ambiguous**       | `[1, 1]`           | 2                        | both shares redeem 0.5 each            |
| **Unresolvable (burn all)** | `[0, 0]`           | 0 (function will revert) | not allowed                            |
